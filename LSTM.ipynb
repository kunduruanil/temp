```python
import pandas as pd
import numpy as np
# Get some time series data
df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/timeseries.csv")
df.head()
```

```python
df.index=df.Date
df.head()
```

```python
df.pop('Date')
df.head()
```

```python
Y=df['G']
df=df.drop(['G'],axis=1)
df.head()
```

```python
df.values
df.shape
```

```python
Y
```

```python
X=[]
y_train=[]
values=df.values
for i in range(len(values)):
    X.append([list(values[i])])
    y_train.append([Y[i]])
```

```python
y_train
```

```python
X
```

```python
df['X']=X
```

```python
x_train=df['X'].cumsum()
```

```python
X[2]
```

```python
x_train[2]
```

```python
import keras
```

```python
from keras.preprocessing.sequence import pad_sequences
x_train_new=pad_sequences(x_train,11)
```

```python
x_train_new.shape
```

```python
x_train[1]
```

```python
x_train_new[1]
```

```python
y_train=np.array(y_train)
y_train
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
help(LSTM)
```

## input layer

```python
s=x_train_new.shape
print(s)
model=Sequential()
model.add(LSTM(10,input_dim=s[2],input_length=s[1]))
```

```python
s
```

## hiden layer

```python
model.add(Dense(10,activation='relu'))
```

## output layer

```python
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

```python
history = model.fit(x_train_new, y_train,batch_size=11, epochs=300)
```

```python
model.predict(x_train_new)
```

```python
y_train
```

```python
model.save("lstm.h5")
```
